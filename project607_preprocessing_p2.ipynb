This means there are strings needing to be converted to ints. See if there are any entries with commas, eg "7,500" rather than "7500"

```python
data.loc[data.votes.str.contains(",")]
```

Replace the commas

```python
data.loc[data.votes.str.contains(","), "votes"] = \
    data.loc[data.votes.str.contains(","), "votes"].apply(lambda x: x.replace(",", ""))
```

See if that worked
