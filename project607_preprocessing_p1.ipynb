### NOTE
Ordinarily, statistics regarding the distribution of empty values would be applied to ascertain which features or samples have too little data to be of value. However, the data is not randomly distributed. It was scraped from IMDb from the top 10000 movies of the year. The most famous movies comprise the upper echelons of the ratings scheme, represented accordingly as the earliest entries scraped for each year. The movies toward the end of the rating scheme are outliers, being obscure, incompletely filled out, and consequently of little verifiable or demonstrable value. Even if the sparse data pertaining to these movies emerged as a highly recommended option, procuring the movie would prove nigh impossible (due to lack of information) and thus render the selection an obsolete one. For this reason, entries missing a majority of the most salient features will be eliminated.
